### Dropping columns with high multicollinearity 

```python
cols_to_drop = [
    "cons_12m",
    "forecast_price_energy_off_peak",
    "forecast_price_energy_peak",
    "latest_price_off_peak_var",
    "latest_price_peak_var",
    "price_peak_diff",
    "price_offpeak_diff",
    "avg_price_diff",
    "tenure_days",
    "contract_age_days",
    "avg_monthly_cons",
    "has_gas_flag",
    "margin_gross_pow_ele",
    "margin_net_pow_ele",
    "latest_price_mid_peak_var",
    "latest_price_mid_peak_fix",
    "latest_price_peak_fix",
    "net_margin",
    "profit_per_product",
    "forecast_cons_year",
    "imp_cons"
]
```

```python
df = df.drop(columns=cols_to_drop)
```

### Feature selection

```python
from sklearn.linear_model import LogisticRegression
import pandas as pd

X = df.drop(columns=['churn'])
y = df['churn']

logit = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=2000
)

logit.fit(X, y)

coef = pd.Series(logit.coef_[0], index=X.columns)

selected_features = coef[coef != 0].index.tolist()

print("Selected Features:", selected_features)
```

```python
print(len(selected_features))
```

```python
df_selected = df[selected_features]
df_selected.head()
```

```python

selected_features = coef[coef != 0].index.tolist()

df_selected = df[selected_features]

# Save to CSV
df_selected.to_csv("selected_features_data.csv", index=False)

print("Saved: selected_features_data.csv")
```
